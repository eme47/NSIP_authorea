####Import data into R. Prune to relevant samples (PS, non-bulk). Get rid of OTUs not observed in at least 25% of samples).

```python
%%R
physeq = import_biom("../data/otu_table_wtax.biom", "../data/otusn.tree")
sample.data = import_qiime_sample_data("../data/barcode_mapper.txt")
physeq = merge_phyloseq(physeq,sample.data)
physeq.snorm = transform_sample_counts(physeq, function(x) x/sum(x))
physeq.thresh = filter_taxa(physeq.snorm, function(x) sum(x > 0) > (0.25 * length(x)), TRUE)
print(physeq)
print(physeq.thresh)
```

####Calculate pairwise weighted unifrac distances between samples.

```python
%%R
registerDoParallel(cores=20)
wunif.dist = distance(physeq.thresh, 
                      method = "unifrac", 
                      weighted = TRUE,
                      fast = TRUE, 
                      parallel = TRUE, 
                      normalized = FALSE)
```

####Ordinate using weighted unifrac distnaces with MDS and plot.

##Ordination of all gradient fraction libraries.

```python
%%R -h 325 -w 650

days = c(2, 4)

df.sites = data.frame()
df.otus = data.frame()

f = function(physeq.obj, day) {
    md = sample_data(physeq.obj)
    p = prune_samples(md$DayCollected == day, physeq.obj)
    mds = ordinate(p, method = "MDS", distance = wunif.dist)
    df.sites = plot_ordination(p, mds, justDF = TRUE)
}

df.sites = ldply(days, f, physeq.obj = physeq.thresh)

p = ggplot(df.sites, aes(x = Axis.1, y = Axis.2, fill = Treatment, size = Density))

p = p + facet_wrap(~DayCollected)

p = p + geom_point(shape = 21)

p = p + scale_size_area(aesthetics = "Density")

p = p + theme_bw()

p = p + guides(fill = guide_legend(override.aes = list(size = 5)))

p = p + labs(x = "PC1", y = "PC2")

p
```

##We can also plot weighted unifrac distance between paired (control versus labeled) gradient fraction libraries against the density of the gradient fraction pair.

```python
%%R
m = as.matrix(wunif.dist)
df.dist = melt(m)[melt(upper.tri(m))$value,]

names(df.dist) = c("X.SampleID", "id2", "distance")

head(df.dist)
```

```python
%%R

md = sample_data(physeq)

m = outer(md$Density, md$Density, "-")
rownames(m) = md$X.SampleID
colnames(m) = md$X.SampleID
df.delta.density = melt(m)[melt(upper.tri(m))$value,]
head(df.delta.density)

md.df = data.frame(md)

f = function(day) {
    d = subset(md.df, DayCollected == day)
    label.ids = subset(d, Treatment == "label")$X.SampleID
    control.ids = subset(d, Treatment == "control")$X.SampleID
    subset(df.delta.density, (X1 %in% label.ids)&(X2 %in% control.ids)) 
}

df.delta.density = ldply(c(2, 4), f)

df.delta.density = ddply(df.delta.density, .(X1), function(x) arrange(x, abs(value))[1,])

names(df.delta.density) = c("X.SampleID", "id2", "delta.density")

head(df.delta.density)
```

```python
%%R
d = join(df.delta.density, df.dist, by = c("X.SampleID","id2"))
d = join(d, data.frame(md)[, c("X.SampleID","Density", "DayCollected")], by = "X.SampleID")
head(d)
```

```python
%%R -h 300 -w 700
p = ggplot(subset(d, abs(delta.density) <= 0.003), 
           aes(x = Density, y = distance, color = abs(delta.density))) 

p = p + facet_wrap(~DayCollected, scale = "free_x")

p = p + geom_smooth(method=lm)

p = p + geom_point(size = 3)

p = p + labs(y = "Weighted Unifrac Distance", color = "density difference, g/mL")

p
```

```python
%%R -h 300 -w 600
p = ggplot(subset(d, abs(delta.density) <= 0.003), 
           aes(x = delta.density, y = distance, color = Density)) 

p = p + facet_wrap(~DayCollected, scale = "free_x")

p = p + geom_point(size = 3)

p = p + labs(y = "Weighted Unifrac Distance", x = "delta density, g/mL")

p
```
